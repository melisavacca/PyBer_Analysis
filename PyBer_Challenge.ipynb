### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data2.csv"
ride_data_to_load = "Resources/ride_data2.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_by_city_type = total_fares_by_city_type / total_rides_by_city_type
average_fare_by_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_by_city_type = total_fares_by_city_type / total_drivers_by_city_type
average_fare_per_driver_by_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pd.DataFrame()

summary_df["Total Rides"] = total_rides_by_city_type
summary_df["Total Drivers"] = total_drivers_by_city_type
summary_df["Total Fares"] = total_fares_by_city_type
summary_df["Average Fare per Ride"] = average_fare_by_city_type
summary_df["Average Fare per Driver"] = average_fare_per_driver_by_city_type

summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
summary_df["Total Rides"] = summary_df["Total Rides"].map("{:,}".format)
summary_df["Total Drivers"] = summary_df["Total Drivers"].map("{:,}".format)
summary_df["Total Fares"] = summary_df["Total Fares"].map("${:,.2f}".format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map("${:.2f}".format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map("${:.2f}".format)

summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_city_type_and_date = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
fare_by_city_type_and_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_by_city_type_and_date = fare_by_city_type_and_date.reset_index()
fare_by_city_type_and_date.tail()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_by_city_type_and_date_pivot = fare_by_city_type_and_date.pivot(index ='date', columns ='type', values ='fare')
fare_by_city_type_and_date_pivot.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_april_df = fare_by_city_type_and_date_pivot.loc['2019-01-01':'2019-04-29']
jan_april_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_april_df.index = pd.to_datetime(jan_april_df.index)
jan_april_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_april_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_april_week = jan_april_df.resample("W").sum()
jan_april_week.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
jan_april_week.plot()

#Add properties for graph
plt.title('Total Fare by City Type')
plt.ylabel('Fare ($USD)')
plt.xlabel('Month')

#Create a legend
plt.legend(title = "City Type")

#Save the figure
plt.savefig("Analysis/PyBer_fare_summary.png")

#Show figure
plt.show()

```
